```python
# ============================
# 1. INSTALAR ULTRALYTICS
# ============================
!pip install ultralytics

from ultralytics import YOLO


# ============================
# 2. MONTAR O GOOGLE DRIVE
# ============================
from google.colab import drive
drive.mount('/content/drive')


# ============================
# 3. CRIAR PASTA DE BACKUP
# ============================
import os

backup_dir = "/content/drive/MyDrive/YOLO_BACKUP"
os.makedirs(backup_dir, exist_ok=True)


# ============================
# 4. CARREGAR MODELO BASE
# ============================
model = YOLO("yolov8s.pt")  # pode trocar para yolov8n.pt, yolov8m.pt etc.


# ============================
# 5. TREINAMENTO OTIMIZADO
# ============================
model.train(
    data="/content/residuos-sólidos-4/data.yaml",  # dataset
    epochs=200,            # mais épocas se dataset for grande
    imgsz=768,             # resolução
    batch=16,              # ajustar conforme GPU
    augment=True,          # aumento de dados
    optimizer='SGD',       # ou 'Adam'
    patience=15,           # early stopping
)


# ============================
# 6. SALVAR BEST.PT E LAST.PT NO DRIVE
# ============================
!cp /content/runs/detect/train/weights/best.pt "/content/drive/MyDrive/YOLO_BACKUP/best.pt"
!cp /content/runs/detect/train/weights/last.pt "/content/drive/MyDrive/YOLO_BACKUP/last.pt"

print("Backup concluído! Arquivos salvos na pasta YOLO_BACKUP no Google Drive.")
```
